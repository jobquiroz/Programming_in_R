> Everything in R is an object.

**Declarando variables**

```
# declaring a variable using `<-`
nameOfVariable <- value

# declaring a variable using `=`
nameOfVariable = value
```


**Tipos de variables**



````r
# Variables of different types

# Real Numeric
myRealNumeric <- 10
class(myRealNumeric)

# Decimal Numeric
myDecimalNumeric <- 10.0
class(myDecimalNumeric)

# Character
myCharacter <- "10"
class(myCharacter)

# Logical
myBoolean <- TRUE
class(myBoolean)

# Integer
myInteger <- 0:10
class(myInteger)

#Complex
myComplex <- 5i
class(myComplex)
````

También se puede usar `typeof()`, pero este actua a un nivel más bajo.

`length()`regresa la longitud de los datos en la variable.

````r
# Testing typeof() and length()
myInteger <- 1:10

cat("The type of the variable is ", typeof(myInteger), "\n") # returns low level data type
cat("The length of the variable is ", length(myInteger), "\n") # returns length of variable
````

**Value, high- and low-level data types**

````r
cat(myRealNumeric, "\n")
cat("The class of the variable is ", class(myRealNumeric), "\n") # returns high level data type
cat("The type of the variable is ", typeof(myRealNumeric), "\n") # returns low level data type
cat("The length of the variable is ", length(myRealNumeric), "\n\n") # returns length of variable

cat(myDecimalNumeric, "\n")
cat("The class of the variable is ", class(myDecimalNumeric), "\n") # returns high level data type
cat("The type of the variable is ", typeof(myDecimalNumeric), "\n") # returns low level data type
cat("The length of the variable is ", length(myDecimalNumeric), "\n\n") # returns length of variable

cat(myCharacter, "\n")
cat("The class of the variable is ", class(myCharacter), "\n") # returns high level data type
cat("The type of the variable is ", typeof(myCharacter), "\n") # returns low level data type
cat("The length of the variable is ", length(myCharacter), "\n\n") # returns length of variable

cat(myBoolean, "\n")
cat("The class of the variable is ", class(myBoolean), "\n") # returns high level data type
cat("The type of the variable is ", typeof(myBoolean), "\n") # returns low level data type
cat("The length of the variable is ", length(myBoolean), "\n\n") # returns length of variable

cat(myInteger, "\n")
cat("The class of the variable is ", class(myInteger), "\n") # returns high level data type
cat("The type of the variable is ", typeof(myInteger), "\n") # returns low level data type
cat("The length of the variable is ", length(myInteger), "\n\n") # returns length of variable

cat(myComplex, "\n")
cat("The class of the variable is ", class(myComplex), "\n") # returns high level data type
cat("The type of the variable is ", typeof(myComplex), "\n") # returns low level data type
cat("The length of the variable is ", length(myComplex), "\n\n") # returns length of variable
````

### Listing variables.

We can check all the variables that have been created in the workspace using the keyword `ls()`.

````r
myRealNumeric <- 10
myDecimalNumeric <- 10.0
myCharacter <- "10"
myBoolean <- TRUE
myInteger <- 0:10
myComplex <- 5i

cat("Variables in the current directory: \n")
ls() # returns all the variables created in the workspace alphabetically
cat("\n")
````

### Deleting

`rm()` removes one or more objects from the workspace.

````r
myRealNumeric <- 10
myDecimalNumeric <- 10.0
myCharacter <- "10"
myBoolean <- TRUE
myInteger <- 0:10
myComplex <- 5i

cat("Variables in the current directory: \n")
ls() # returns all the variables created in the workspace
cat("\n")

cat("Deleting myRealNumeric and myDecimalNumeric \n\n")

rm(myRealNumeric, myDecimalNumeric) # delete the two mentioned variables

cat("Variables in the current directory, now: \n")
ls() # returns all the variables created in the workspace
     # myRealNumeric, myDecimalNumeric are now deleted
cat("\n")
````

### Strings

`cat` can be used with single quotes or with doubles:

````r
cat("learning is fun \n")
cat('learning is fun')
````

**Empty String**

````r
cat("") # o cat('')
````

**Escape Sequences**

Empieza con `\` y se usa para incluir caracteres especiales.



Ejemplos

````r
cat("Learning\nis\nfun")
````

````r
cat("Learning\tis\tfun")
````

````r
cat("Learning\\is\\fun")
````

### Operaciones básicas con cadenas

**nchar**

````r
nchar("learning is fun\n")   # espacios, tabs y next line son también incluidos en la cuenta.
````

`length` aquí no se usa, porque toda la cadena se considera un sólo elemento.

````r
length("learning is fun\n")  
````

**Concatenación**

`paste()`

````r
paste("learning", "is", "fun")
````

Por default se utiliza el espacio como separador `sep`.
Pero esto se puede modificar:

````r
paste("learning", "is", "fun", sep = ".")
````

````r
paste("learning", "is", "fun", sep = "")
````

**Duplicando cadenas**

`replicate()`:
 
 > `replicate(<numberOfTimes>, <stringToReplicate>)`
 
y `rep`:

> `rep(<stringToReplicate>, <numberOfTimes>)`

````r
replicate(3, "learning is fun")
````

````r
rep("learning is fun", 3)
````
